```julia
using DataFrames
using Chemfiles
using Gadfly
using Compose
using Cairo
set_default_plot_size(30cm, 12cm)
home = "/home/german/labo/18/egfr/model/"
```

```julia
tmp = readdlm(string(home, "/pdbs.list"))
pdbs_list = convert(Array{String, 1}, reshape(tmp, length(tmp)))

best_models = Array{Int64, 1}(length(pdbs_list))
peor_aa = Array{Float64, 1}(length(pdbs_list))

i = 0
for each in pdbs_list
    i += 1
    println(each)

    energías = zeros(277, 50)
    for x = 1:50
        modelo = String
        if x < 10
            modelo = string(each, "_full.B9999000", x, ".pdb")
        else
            modelo = string(each, "_full.B999900", x, ".pdb")
        end
        perfil = string("profile_", modelo[1:end-4])
    
        try 
            energías[:, x] = readdlm(string(home, each, "/", perfil))[:, end]
        catch
            println("\tFailed model: ", x)
        end
    end

    best_models[i] = indmin(mapslices(x -> mean(x), energías, 1))
    peor_aa[i] = maximum(mapslices(x -> mean(x), energías, 1))
end
```

```julia
df_energías = DataFrame(DOPE = reshape(energías, size(energías)[1] * size(energías)[2]),
    Iteration = repeat(collect(1:50), inner = [277]), 
    Residue = repeat(collect(1:277), outer = [50]));
```

```julia
pdf_T_lig = plot(df_energías, Geom.line, x="Residue", y="DOPE", color = "Iteration",

    #Guide.title("NEQ run (1-10)ps. Ligand Temperature"),
    #Coord.cartesian(ymin = 400, ymax = 1200, xmax = 10, xmin = 1),
    #Guide.yticks(ticks = collect(400:200:1200)),

    Theme(line_width = 1pt, grid_line_width = 1pt,
        minor_label_font_size = 24pt, major_label_font_size = 30pt, key_label_font_size = 18pt,
        key_title_font_size = 22pt))
```
