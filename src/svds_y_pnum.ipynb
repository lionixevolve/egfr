```julia
function tognm(vtor_anm)
    vtor_gnm = Array{Float64}(convert(Int64, length(vtor_anm)/3));
    vtor_anm =  vtor_anm.^2
    for i=1:convert(Int64, length(vtor_anm)/3)
        vtor_gnm[i] = sqrt(vtor_anm[i*3-2] + vtor_anm[i*3-1] + vtor_anm[i*3])
    end
    return vtor_gnm
end
```

```julia
function WeightedHist(in_vec::Array{Float64, 1}, in_bins, in_weight, density = false)
    # Safety check    
    if length(in_vec) != length(in_weight)
        println("Each element of the input vector needs one weight")
        return
    end
    
    # Prepare variables
    out_counts = Array{Float64}(length(in_bins)-1)
    
    # Get weighted histogram
    for i=1:length(in_bins)-1
        if i == 1
            # Include those that fall before the beggining of the bins
            temp_bool = (in_vec .>= in_bins[i]) .& (in_vec .< in_bins[i+1]) .| (in_vec .<= in_bins[i])
            out_counts[i] = sum(in_weight[temp_bool])
        elseif i == length(in_bins)-1
            # Include those that fall after the end of the bins
            temp_bool = (in_vec .>= in_bins[i]) .& (in_vec .< in_bins[i+1]) .| (in_vec .>= in_bins[end])
            out_counts[i] = sum(in_weight[temp_bool])
        else
            temp_bool = (in_vec .>= in_bins[i]) .& (in_vec .< in_bins[i+1])
            out_counts[i] = sum(in_weight[temp_bool])
        end
    end
    
    # Get bins middle points
    out_middle = (in_bins[1:end-1] + in_bins[2:end]) / 2
    
    # Turn counts into density
    if (density == true)
        out_counts = out_counts ./ sum(out_counts) 
    end
    return out_counts, out_middle
end
```

```julia
# IM
ref_im = 9
todos_im = [ modos_im_1 , modos_im_2 , modos_im_3 , modos_im_4 , modos_im_5 ]

# 1
new_modos_im_1 = Array{Float64, 2}(aa_3, length(im_list))
new_modos_im_1[:, ref_im] = modos_im_1[:, ref_im]
for i = 1:length(im_list)
    # Determino q modo de la struct actual tiene mayor solapimiento con el modo 1 de 3W32_A (referencia)
    tmp = abs.([ dot(modos_im_1[:, 9], modos_im_1[:, i]) ; dot(modos_im_1[:, ref_im], modos_im_2[:, i]) ;
        dot(modos_im_1[:, ref_im], modos_im_3[:, i]) ; dot(modos_im_1[:, ref_im], modos_im_4[:, i]) ; 
        dot(modos_im_1[:, ref_im], modos_im_5[:, i]) ])

    # Pongo ese modo como modo 1 de la struct actual
    new_modos_im_1[:, i] = todos_im[indmax(tmp)][:, i]
end

# 2
new_modos_im_2 = Array{Float64, 2}(aa_3, length(im_list))
new_modos_im_2[:, ref_im] = modos_im_2[:, ref_im]
for i = 1:length(im_list)
    # Determino q modo de la struct actual tiene mayor solapimiento con el modo 1 de 3W32_A (referencia)
    tmp = abs.([ dot(modos_im_2[:, 9], modos_im_1[:, i]) ; dot(modos_im_2[:, ref_im], modos_im_2[:, i]) ;
        dot(modos_im_2[:, ref_im], modos_im_3[:, i]) ; dot(modos_im_2[:, ref_im], modos_im_4[:, i]) ; 
        dot(modos_im_2[:, ref_im], modos_im_5[:, i]) ])

    # Pongo ese modo como modo 1 de la struct actual
    new_modos_im_2[:, i] = todos_im[indmax(tmp)][:, i]
end


# AM
ref_am = 1
todos_am = [ modos_am_1 , modos_am_2 , modos_am_3 , modos_am_4 , modos_am_5 ]

# 1
new_modos_am_1 = Array{Float64, 2}(aa_3, length(am_list))
new_modos_am_1[:, ref_am] = modos_am_1[:, ref_am]
for i = 2:length(am_list)
    # Determino q modo de la struct actual tiene mayor solapamiento con el modo 1 de 1M14_A (referencia)
    tmp = abs.([ dot(modos_am_1[:, ref_am], modos_am_1[:, i]) ; dot(modos_am_1[:, ref_am], modos_am_2[:, i]) ;
        dot(modos_am_1[:, ref_am], modos_am_3[:, i]) ; dot(modos_am_1[:, ref_am], modos_am_4[:, i]) ; 
        dot(modos_am_1[:, ref_am], modos_am_5[:, i]) ])

    # Pongo ese modo como modo 1 de la struct actual
    new_modos_am_1[:, i] = todos_am[indmax(tmp)][:, i]
end

# 2
new_modos_am_2 = Array{Float64, 2}(aa_3, length(am_list))
new_modos_am_2[:, ref_am] = modos_am_2[:, ref_am]
for i = 2:length(am_list)
    # Determino q modo de la struct actual tiene mayor solapamiento con el modo 1 de 1M14_A (referencia)
    tmp = abs.([ dot(modos_am_2[:, ref_am], modos_am_1[:, i]) ; dot(modos_am_2[:, ref_am], modos_am_2[:, i]) ;
        dot(modos_am_2[:, ref_am], modos_am_3[:, i]) ; dot(modos_am_2[:, ref_am], modos_am_4[:, i]) ; 
        dot(modos_am_2[:, ref_am], modos_am_5[:, i]) ])

    # Pongo ese modo como modo 1 de la struct actual
    new_modos_am_2[:, i] = todos_am[indmax(tmp)][:, i]
end
```

### Obtengo pnumbers de colectividad

```julia
# IM
gnm_new_modos_im_1 = mapslices(x->tognm(x), new_modos_im_1, 1)
tmp = transpose((mapslices(x->sum(x), (gnm_new_modos_im_1 .^ 4), 1)) .^ -1) ./ aa
pnu_new_modos_im_1 = reshape(tmp, length(tmp))
# 2
gnm_new_modos_im_2 = mapslices(x->tognm(x), new_modos_im_2, 1)
tmp = transpose((mapslices(x->sum(x), (gnm_new_modos_im_2 .^ 4), 1)) .^ -1) ./ aa
pnu_new_modos_im_2 = reshape(tmp, length(tmp))

# AM
gnm_new_modos_am_1 = mapslices(x->tognm(x), new_modos_am_1, 1)
tmp = transpose((mapslices(x->sum(x), (gnm_new_modos_am_1 .^ 4), 1)) .^ -1) ./ aa
pnu_new_modos_am_1 = reshape(tmp, length(tmp))
# 2
gnm_new_modos_am_2 = mapslices(x->tognm(x), new_modos_am_2, 1)
tmp = transpose((mapslices(x->sum(x), (gnm_new_modos_am_2 .^ 4), 1)) .^ -1) ./ aa;
pnu_new_modos_am_2 = reshape(tmp, length(tmp));
```
